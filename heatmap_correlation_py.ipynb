Dados retirados do link do Kaggle: 
https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data?select=train.csv

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Lendo o Arquivo dos dados
df = pd.read_csv('/content/train.csv')

# Definindo somente as colunas desejadas
df = df[['OverallQual', 'GarageYrBlt', 'KitchenAbvGr','LotArea', 'YearBuilt', 'MoSold', 'YrSold', 'SalePrice']]

# Renomeando as Colunas
df.rename(columns={'OverallQual': 'Qualidade_Geral', 'GarageYrBlt':'Qualidade_Garagem',\
                   'KitchenAbvGr':'Qualidade_Cozinha','LotArea':'Area_total',\
                   'YearBuilt': 'Ano_Constr','MoSold':'Mês_venda', 'YrSold': 'Ano_venda',\
                   'SalePrice':'Valor_venda'}, inplace=True)
# Matriz de Correlação
df.corr()

#Definindo a area de impressão
plt.figure(figsize=(10, 6))

# Definindo o tipo do gráfico, valor minimo e maximo, e apresentação dos dados
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
# Nomeando o Mapa e ajustando a fonte
heatmap.set_title('Mapa de Calor de Correlação', fontdict={'fontsize':18}, pad=10)

# Salvando a imagem no diretorio
plt.savefig('/content/heatmap_corr.png', dpi=300, bbox_inches='tight')
```
